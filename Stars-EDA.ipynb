```python
#Creating an extra column to aggregate The Stars more efficently
#Morethan 10000 Light years-Farthest Stars
#lessthan 500 Light years-Medium Farthest
#Lessthan 100 Faraway stars
#less than 50- NearStars
#less than 10-NearestStar
```

```python
def star_distance_group(Distance):
    if(Distance>1000):
        return 'Distant Stars(1000Ly+)'
    elif (500<Distance<=1000):
        return 'Far Stars(500ly-)'
    elif (100 <Distance<=500):
        return 'Mid Stars(500ly-)'
    elif (50<Distance<=100):
        return 'Close Stars(100Ly-)'
    elif(10<Distance<=50):
        return 'Very close Stars(50Ly-)'
    else:
        return 'Nearby star(10Ly-)'
        
```

```python
Stars['Distance group']=Stars['Distance (ly)'].apply(star_distance_group)
```

```python
Stars
```

```python
X=Stars['Distance group'].value_counts(normalize='True')
plt.pie(X.values,labels=X.index,autopct='%1.01f%%',explode=[0,0,0,0.025,0.05,0.075])
```

```python
#Creating a grouping in Star Radius to divide it
#Super Giants: >700
#Giants: >400
#Large Stars: >100
#Medium Stars: 50â€“100
#Small Stars: 10â€“50
#Dwarfs: <1
```

```python
def star_radius_grp(radius):
    if(radius>700):
        return 'Super Gaints'
    elif (400<radius<=700):
        return 'Gaints'
    elif(100<radius<=400):
        return 'Large Stars'
    elif(50<radius<=100):
        return 'Medium Star'
    elif(10<radius<=50):
        return 'Small Stars'
    else:
        return 'dwarf Stars'
```

```python
Stars['Size Group']=Stars['Radius (R/Ro)'].apply(star_radius_grp)
```
